```python
import pandas as pd


def sort_month(df, column_name):
    # Определяем порядок месяцев
    months_order = [
        'January', 'February', 'March', 'April', 'May', 'June',
        'July', 'August', 'September', 'October', 'November', 'December'
    ]

    # Создаём категориальный тип
    df[column_name] = pd.Categorical(
        df[column_name],
        categories=months_order,
        ordered=True
    )

    # Сортируем и сбрасываем индекс
    return df.sort_values(column_name).reset_index(drop=True)


# Пример использования
sorted_df = sort_month(df, 'arrival_date_month')
```

```python
from sklearn.model_selection import train_test_split

# Разделение на Train / Validation / Test

# тестовая
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# валидационная
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.20, random_state=42
)

print(f"Train shape:      {X_train.shape}")
print(f"Validation shape: {X_val.shape}")
print(f"Test shape:       {X_test.shape}")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.columns.tolist())
    ]
)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline

knn_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', KNeighborsRegressor(n_neighbors=5))
])
```

# kNN

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import itertools

# Параметры для перебора
n_neighbors_list = [3, 5, 7, 9, 11, 15, 20]
weights_list = ['uniform', 'distance']
p_list = [1, 2]  # 1 = Manhattan, 2 = Euclidean

print(f"{'n_neighbors':>6} | {'weights':>10} | {'p':>3} | {'MAE':>8} | {'RMSE':>8} | {'R²':>8}")
print("-" * 70)

best_score = -np.inf
best_params = None
results = []

for k, w, p in itertools.product(n_neighbors_list, weights_list, p_list):
    knn_pipeline.set_params(
        regressor__n_neighbors=k,
        regressor__weights=w,
        regressor__p=p
    )

    knn_pipeline.fit(X_train, y_train)
    y_pred_val = knn_pipeline.predict(X_val)

    mae = mean_absolute_error(y_val, y_pred_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred_val))
    r2 = r2_score(y_val, y_pred_val)

    print(f"{k:6} | {w:10} | {p:3} | {mae:8.3f} | {rmse:8.3f} | {r2:8.4f}")

    # Сохраняем лучший результат по R²
    if r2 > best_score:
        best_score = r2
        best_params = (k, w, p)
        best_mae = mae
        best_rmse = rmse

print("\n" + "=" * 70)
print(f"Лучшая комбинация параметров:")
print(f"n_neighbors = {best_params[0]}, weights = {best_params[1]}, p = {best_params[2]}")
print(f"MAE  = {best_mae:.3f} | RMSE = {best_rmse:.3f} | R² = {best_score:.4f}")
```

* MAE — Mean Absolute Error (Средняя абсолютная ошибка)

Это среднее значение, на сколько в среднем модель ошибается в предсказании цены за ночь (adr).

* RMSE — Root Mean Squared Error (Корень из среднеквадратичной ошибки

Это тоже средняя ошибка, но со штрафом за большие отклонения.

* R² — Коэффициент детерминации (R-squared)
Показывает, какую долю изменчивости целевой переменной (adr) модель смогла объяснить.

R² = 1.0 → модель идеально объясняет все изменения ADR
R² = 0.69 → модель объясняет 69% вариации цены за ночь
R² = 0 → модель работает не лучше, чем простое среднее значение
R² < 0 → модель хуже, чем просто предсказывать среднее

```python
best_params = {'n_neighbors': 11, 'weights': 'distance', 'p': 1}

knn_pipeline.set_params(
    regressor__n_neighbors=best_params['n_neighbors'],
    regressor__weights=best_params['weights'],
    regressor__p=best_params['p']
)

# Обучаем на всей доступной обучающей выборке
knn_pipeline.fit(X_train_full, y_train_full)

# Предсказываем на тестовой выборке
y_pred_test = knn_pipeline.predict(X_test)

# Считаем метрики
mae_test = mean_absolute_error(y_test, y_pred_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2_test = r2_score(y_test, y_pred_test)

print("KNN Regressor — Результаты на ТЕСТОВОЙ выборке")
print(f"Параметры: n_neighbors={best_params['n_neighbors']}, "
      f"weights='{best_params['weights']}', p={best_params['p']}")
print(f"MAE  = {mae_test:.3f}")
print(f"RMSE = {rmse_test:.3f}")
print(f"R²   = {r2_test:.4f}")
```

```python
param_grid_log = {
    'classifier__alpha': [0.0001, 0.001, 0.01],           # сила регуляризации
    'classifier__learning_rate': ['optimal', 'adaptive'], # стратегия обучения
    'classifier__penalty': ['l2', 'l1', 'elasticnet'],    # тип регуляризации
    'classifier__eta0': [0.01, 0.1]                        # начальная скорость (для adaptive)
}
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import numpy as np
import time

print("Начинаем перебор параметров для Logistic Regression (SGD + PCA)...\n")
print(f"{'alpha':>8} | {'learning_rate':>13} | {'penalty':>10} | {'eta0':>6} | {'Acc':>8} | {'F1':>8} | Time")
print("-" * 90)

best_f1 = -np.inf
best_params_log = None
start_total = time.time()

for alpha in param_grid_log['classifier__alpha']:
    for lr in param_grid_log['classifier__learning_rate']:
        for penalty in param_grid_log['classifier__penalty']:
            for eta0 in param_grid_log['classifier__eta0']:

                start = time.time()

                log_reg_pipeline.set_params(
                    classifier__alpha=alpha,
                    classifier__learning_rate=lr,
                    classifier__penalty=penalty,
                    classifier__eta0=eta0
                )

                # Обучаем на Xcl_train (классификация)
                log_reg_pipeline.fit(Xcl_train, ycl_train)
                ycl_pred_val = log_reg_pipeline.predict(Xcl_val)

                acc = accuracy_score(ycl_val, ycl_pred_val)
                f1 = f1_score(ycl_val, ycl_pred_val)
                prec = precision_score(ycl_val, y_pred_val)
                rec = recall_score(ycl_val, y_pred_val)


                elapsed = time.time() - start

                print(f"{alpha:8} | {lr:13} | {penalty:10} | {eta0:6} | {acc:8.3f} | {f1:8.3f} | {elapsed:.1f}s")

                if f1 > best_f1:
                    best_f1 = f1
                    best_params_log = (alpha, lr, penalty, eta0)
                    best_acc = acc
                    best_prec = prec
                    best_rec = rec

print("\n" + "="*90)
print(f"Лучшая комбинация параметров для Logistic Regression:")
print(f"alpha={best_params_log[0]}, learning_rate={best_params_log[1]}, penalty={best_params_log[2]}, eta0={best_params_log[3]}")
print(f"Accuracy = {best_acc:.4f} | Precision = {best_prec:.4f} | Recall = {best_rec:.4f} | F1-Score = {best_f1:.4f}")
print(f"Общее время: {(time.time() - start_total):.1f} секунд")
```

Accuracy (Acc): Показывает общую долю правильных предсказаний (сколько раз мы угадали статус "отмена/не отмена").

F1-Score (F1): Это "золотая середина" между точностью и полнотой. В отельном бизнесе важно не просто угадывать, а находить как можно больше реальных отмен, не ошибаясь при этом слишком часто.

# Градиетный бустинг

```python
from xgboost import XGBClassifier

xgb_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    # ('pca', PCA(n_components=10)),
    ('classifier', XGBClassifier(random_state=42, eval_metric='logloss'))
])
```

```python
param_grid_xgb = {
    'classifier__n_estimators': [50, 100],        # количество деревьев
    'classifier__learning_rate': [0.01, 0.1],     # скорость обучения (шаг)
    'classifier__max_depth': [3, 5, 7],           # глубина деревьев
    'classifier__subsample': [0.8, 1.0]           # доля данных для обучения одного дерева
}
```

```python
from sklearn.metrics import accuracy_score, f1_score
import time


print("Начинаем перебор параметров для XGBoost...\n")
print(f"{'n_est':>5} | {'LR':>5} | {'depth':>5} | {'sub':>5} | {'Acc':>8} | {'F1':>8} | Time")
print("-" * 75)

best_f1_xgb = -np.inf
best_params_xgb = None
start_total = time.time()

for n_est in param_grid_xgb['classifier__n_estimators']:
    for lr in param_grid_xgb['classifier__learning_rate']:
        for depth in param_grid_xgb['classifier__max_depth']:
            for sub in param_grid_xgb['classifier__subsample']:

                start = time.time()

                xgb_pipeline.set_params(
                    classifier__n_estimators=n_est,
                    classifier__learning_rate=lr,
                    classifier__max_depth=depth,
                    classifier__subsample=sub
                )

                # Обучаем на сбалансированных данных
                xgb_pipeline.fit(X_resampled, y_resampled)
                y_pred_val = xgb_pipeline.predict(Xcl_val)

                acc = accuracy_score(ycl_val, y_pred_val)
                prec = precision_score(ycl_val, y_pred_val)
                rec = recall_score(ycl_val, y_pred_val)
                f1 = f1_score(ycl_val, y_pred_val)
                elapsed = time.time() - start

                print(f"{n_est:5} | {lr:5} | {depth:5} | {sub:5} | {acc:8.3f} | {prec:8.3f} | {rec:8.3f} | {f1:8.3f} | {elapsed:.1f}s")

                if f1 > best_f1_xgb:
                    best_f1_xgb = f1
                    best_params_xgb = (n_est, lr, depth, sub)
                    best_acc_xgb = acc
                    best_prec_xgb = prec
                    best_rec_xgb = rec

print("\n" + "="*75)
print(f"Лучшая комбинация: n_estimators={best_params_xgb[0]}, LR={best_params_xgb[1]}, Depth={best_params_xgb[2]}")
print(f"Accuracy = {best_acc_xgb:.4f} | Precision = {best_prec_xgb:.4f} | Recall = {best_rec_xgb:.4f} | F1-Score = {best_f1_xgb:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score, f1_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# 1. Применяем лучшие параметры именно к XGB_PIPELINE
xgb_pipeline.set_params(
    classifier__n_estimators=best_params_xgb[0],
    classifier__learning_rate=best_params_xgb[1],
    classifier__max_depth=best_params_xgb[2],
    classifier__subsample=best_params_xgb[3]
)

xgb_pipeline.fit(X_resampled, y_resampled)
ycl_pred_test_xgb = xgb_pipeline.predict(Xcl_test)

print(f"XGBoost — Результаты на ТЕСТОВОЙ выборке:")
print(f"Accuracy = {accuracy_score(ycl_test, ycl_pred_test_xgb):.4f}")
print(f"Precision = {precision_score(ycl_test, ycl_pred_test_xgb):.4f}")
print(f"Recall = {recall_score(ycl_test, ycl_pred_test_xgb):.4f}")
print(f"F1-Score = {f1_score(ycl_test, ycl_pred_test_xgb):.4f}")

print("\nДетальный отчет (Classification Report):")
print(classification_report(ycl_test, ycl_pred_test_xgb))
```
