**Перевод**

hotel — отель;

is_canceled — отменён (бронирование);

lead_time — время до заезда (количество дней между датой бронирования и датой заезда);

arrival_date_year — год заезда;

arrival_date_month — месяц заезда;

arrival_date_week_number — номер недели заезда;

arrival_date_day_of_month — день месяца заезда;

stays_in_weekend_nights — количество ночей, проведённых в выходные;

stays_in_week_nights — количество ночей, проведённых в будние дни;

adults — количество взрослых;

children — количество детей;

babies — количество младенцев;

meal — тип питания;

country — страна (происхождения гостя);

market_segment — сегмент рынка (категория клиентов);

distribution_channel — канал распределения (способ бронирования: онлайн‑турагент, офлайн‑турагент и т. д.);

is_repeated_guest — повторный гость (признак того, что гость уже останавливался в отеле);

previous_cancellations — количество предыдущих отмен бронирований;

previous_bookings_not_canceled — количество предыдущих не отменённых бронирований;

reserved_room_type — зарезервированный тип номера;

assigned_room_type — присвоенный (фактический) тип номера;

booking_changes — количество изменений в бронировании;

deposit_type — тип депозита (гарантии бронирования);

agent — агент (идентификатор турагента или агентства);

company — компания (идентификатор компании, оплачивающей проживание);

days_in_waiting_list — количество дней в листе ожидания;

customer_type — тип клиента;

adr — средняя дневная ставка (Average Daily Rate, цена за ночь);

required_car_parking_spaces — количество требуемых мест на парковке;

total_of_special_requests — общее количество специальных запросов;

reservation_status — статус бронирования;

reservation_status_date — дата статуса бронирования.

```python
import pandas as pd

def sort_month(df, column_name):
    # Определяем порядок месяцев
    months_order = [
        'January', 'February', 'March', 'April', 'May', 'June',
        'July', 'August', 'September', 'October', 'November', 'December'
    ]

    # Создаём категориальный тип
    df[column_name] = pd.Categorical(
        df[column_name],
        categories=months_order,
        ordered=True
    )

    # Сортируем и сбрасываем индекс
    return df.sort_values(column_name).reset_index(drop=True)

# Пример использования
sorted_df = sort_month(df, 'arrival_date_month')
```

```python
final_prices = sort_month(final_hotel, 'month')
final_prices
```

```python
cat_cols = [
    col for col in df.columns
    if df[col].dtype in ['O', 'string', 'category']
]
print(cat_cols)
```

```python
cat_df = df[cat_cols]
cat_df.head()
```

```python
cat_df['reservation_status_date'] = pd.to_datetime(cat_df['reservation_status_date'])

cat_df['year'] = cat_df['reservation_status_date'].dt.year
cat_df['month'] = cat_df['reservation_status_date'].dt.month
cat_df['day'] = cat_df['reservation_status_date'].dt.day
```

```python
cat_df.drop(['reservation_status_date','arrival_date_month'] , axis = 1, inplace = True)
```

```python
cat_df.head()
```

```python
# уникальные значения
for col in cat_df.columns:
    print(f"{col}: \n{cat_df[col].unique()}\n")
```

```python
# кодируем категориальные

cat_df['hotel'] = cat_df['hotel'].map({'Resort Hotel' : 0, 'City Hotel' : 1})

cat_df['meal'] = cat_df['meal'].map({'BB' : 0, 'FB': 1, 'HB': 2, 'SC': 3, 'Undefined': 4})

cat_df['market_segment'] = cat_df['market_segment'].map({'Direct': 0, 'Corporate': 1, 'Online TA': 2, 'Offline TA/TO': 3,
                                                           'Complementary': 4, 'Groups': 5, 'Undefined': 6, 'Aviation': 7})

cat_df['distribution_channel'] = cat_df['distribution_channel'].map({'Direct': 0, 'Corporate': 1, 'TA/TO': 2, 'Undefined': 3,
                                                                       'GDS': 4})

cat_df['reserved_room_type'] = cat_df['reserved_room_type'].map({'C': 0, 'A': 1, 'D': 2, 'E': 3, 'G': 4, 'F': 5, 'H': 6,
                                                                   'L': 7, 'B': 8})

cat_df['deposit_type'] = cat_df['deposit_type'].map({'No Deposit': 0, 'Refundable': 1, 'Non Refund': 3})

cat_df['customer_type'] = cat_df['customer_type'].map({'Transient': 0, 'Contract': 1, 'Transient-Party': 2, 'Group': 3})

cat_df['year'] = cat_df['year'].map({2015: 0, 2014: 1, 2016: 2, 2017: 3})
```

```python
num_df = df.drop(columns = cat_cols)
num_df.drop('is_canceled', axis = 1, inplace = True)
num_df
```

```python
import numpy as np

num_df['lead_time'] = np.log(num_df['lead_time'] + 1)
num_df['arrival_date_week_number'] = np.log(num_df['arrival_date_week_number'] + 1)
num_df['arrival_date_day_of_month'] = np.log(num_df['arrival_date_day_of_month'] + 1)
num_df['agent'] = np.log(num_df['agent'] + 1)
num_df['adr'] = np.log(num_df['adr'] + 1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Выбираем признаки для визуализации (можете заменить на список интересующих столбцов из X)
features_to_plot = ['lead_time', 'adr', 'adults', 'hotel', 'distribution_channel', 'market_segment', 'deposit_type']

# Ограничиваем количество признаков на графике (если их слишком много)
features_to_plot = features_to_plot[:8]  # берём первые 8 признаков для примера

# Создаём фигуру с подзаграфами (4 строки, 2 столбца)
plt.figure(figsize=(15, 20))

# Объединяем X и y для удобства визуализации
data_with_target = pd.concat([X[features_to_plot], y], axis=1)
data_with_target.columns = list(features_to_plot) + ['is_canceled']

# Рисуем графики
for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(4, 2, i)
    sns.histplot(
        data=data_with_target,
        x=feature,
        hue='is_canceled',  # разбивка по целевому признаку (0 и 1)
        kde=True,           # добавляем оценку плотности (кривую KDE)
        bins=30,           # количество бинов
        alpha=0.7          # прозрачность
    )
    plt.title(f'Распределение {feature} по статусу отмены бронирования')
    plt.xlabel(feature)
    plt.ylabel('Частота')

# Настраиваем расположение графиков
plt.tight_layout()
plt.show()
```
